# Introduction

In this sales analysis project, we aim to uncover valuable insights from sales data to support business decision-making. This analysis will address several key questions that can help improve sales strategies, inventory management, and overall business performance. The primary objectives of this analysis include:

1. Understanding Sales Trends: Identifying the best months for sales to help in planning and forecasting.
2. Geographical Insights: Determining which cities generate the maximum number of orders to optimize regional marketing efforts.

4. Sales Patterns: Investigating trends for the most sold products to anticipate future demand.

By addressing these questions, this project will provide a comprehensive view of the sales landscape and offer actionable insights to enhance business strategies.

```python
products = count_df['Product'].values
```

```python
fig , ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1.bar(count_df['Product'] , count_df['Quantity Ordered'])
ax1.set_xticklabels(products , rotation = 'vertical' , fontsize = 8)

ax1.set_ylabel('Order Count')
```

```python
fig , ax1 = plt.subplots()

ax2 = ax1.twinx()

ax1.bar(count_df['Product'] , count_df['Quantity Ordered'] , color = 'r')
ax1.set_xticklabels(products , rotation = 'vertical' , fontsize = 8)

ax2.plot(count_df['Product'] , count_df['Price Each'] , color = 'black')


ax1.set_ylabel('Order Count')
ax2.set_ylabel('Avg price of product')
```

```python
df_duplicated = all_data[all_data['Order ID'].duplicated(keep=False)]
```

```python
df_duplicated
```

```python
df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x : ','.join(x))
```

```python
df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x : ','.join(x)).reset_index()
```

```python
df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x : ','.join(x)).reset_index().rename(columns={'Product' : 'grouped_products'})
```

```python
dup_products = df_duplicated.groupby(['Order ID'])['Product'].apply(lambda x : ','.join(x)).reset_index().rename(columns={'Product' : 'grouped_products'})
```

```python
dup_products
```

```python
df_duplicated.merge(dup_products , how = 'left' , on = 'Order ID')
```

```python
dup_products_df = df_duplicated.merge(dup_products , how = 'left' , on = 'Order ID')
```

```python
dup_products_df
```

```python
dup_products_df.drop_duplicates(subset=['Order ID'])
```

```python
no_dup_df = dup_products_df.drop_duplicates(subset=['Order ID'])
```
